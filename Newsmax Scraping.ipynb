```python
links_2019 = ['https://www.newsmax.com/archives/politics/1/2019/2/','https://www.newsmax.com/archives/politics/1/2019/1/']

links_2018 = [ 'https://www.newsmax.com/archives/politics/1/2018/'
              + str(x) + '/'
              for x in range(1,13) ]
```

```python
newsmax_links = links_2019 + links_2018
```

```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
# from selenium.webdriver.common.keys import Keys
# from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait # available since 2.4.0
from selenium.webdriver.support import expected_conditions as EC # available since 2.26.0
from selenium.webdriver.firefox.options import Options as selenium_options

import os

from fake_useragent import UserAgent

import time
import re
import random
import numpy as np
import pandas as pd
import pickle

ua = UserAgent()

ffdriver = "/Applications/Firefox.app/" # path to the chromedriver executable
os.environ["webdriver.Firefox.driver"] = ffdriver
browser_options = selenium_options()
browser_options.add_argument('--headless')
```

```python
def extractNewsmaxArticleUrls(link):
    driver = webdriver.Firefox(ffdriver, options=browser_options)
    driver.get(link)
    
    elems = driver.find_elements_by_xpath("//*[@href]")

    z = []
    
    for elem in elems:
        href = elem.get_attribute('href')
        
        if 'https://www.newsmax.com/politics/' in href:
            z.append(href)
    
    timeDelay = random.randrange(8000, 12000)/1000
    time.sleep(timeDelay) 
    
    driver.quit()
    
    print("Scraped " + str(link)) 
    print("Waited " + str(timeDelay))
    
    return list(set(z))
```

```python
def allPageLinks(lst_of_links):
    allLinks = []
    for x in lst_of_links:
        allLinks.append(extractNewsmaxArticleUrls(x))
    return np.array(allLinks).flatten()
```

```python
newsmax_article_links = [ extractNewsmaxArticleUrls(link) for link in newsmax_links ]
```

```python
newsmax_article_links
```

```python
def flattener(list_of_lists):    
    flattened_array = []

    for x in list_of_lists:
        for y in x:
            flattened_array.append(y)
    return flattened_array
```

```python
newsmax_article_links_flattened = flattener(newsmax_article_links)
```

```python
with open('newsmax_article_links_flattened.pkl', 'wb') as f:
    pickle.dump(newsmax_article_links_flattened, f)
```

```python
def scrapeNewsmaxArticles(lst_of_articles):
    headers = {'User-Agent': ua.random}
    article_obj = {}
    
    for idx, link in enumerate(lst_of_articles):
        response = requests.get(link, headers)
        soup = BeautifulSoup(response.text)
        
        try:
            date_time = soup.find(class_="artPgDate").text
        except:
            date_time = None
            
        try:
            article_title = soup.find("h1").text
        except:
            article_title = None
        
        try:
            author = soup.find(class_="artPgByline").text
        except:
            author = None
            
        try:
            article_site_topic = 'Politics'
        except:
            article_site_topic = None
            
        try:
            article_body = soup.find(class_="artPgMnStryWrapper").text
            article_body = article_body[:article_body.find('Related Stories')]        
        except:
            article_body = None
            
        article_obj[idx] = [date_time, article_title,author, article_site_topic, article_body]
        
#         timeDelay = random.randrange(2000, 5000)/1000
#         time.sleep(timeDelay) 

        print("Scraped " + str(link)) 
#         print("Waited " + str(timeDelay))    
        
    return article_obj
```

```python
data = scrapeNewsmaxArticles(newsmax_article_links_flattened)
```

```python
with open('newsmax_data.pkl', 'wb') as f:
    pickle.dump(data, f)
```
