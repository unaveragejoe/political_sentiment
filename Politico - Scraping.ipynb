```python
def extractPoliticoArticleUrls(link):
    driver = webdriver.Firefox(ffdriver, options=browser_options)
    driver.get(link)
    
    elems = driver.find_elements_by_xpath("//*[@href]")

    z = []
    
    for elem in elems:
        href = elem.get_attribute('href')
        
        if 'https://www.politico.com/story/' in href:
            z.append(href)
    
#     timeDelay = random.randrange(8000, 12000)/1000
#     time.sleep(timeDelay) 
    
    driver.quit()
    
    print("Scraped " + str(link)) 
#     print("Waited " + str(timeDelay))
    
    return list(set(z))
```

```python
def allPageLinks(lst_of_links):
    allLinks = []
    for x in lst_of_links:
        allLinks.append(extractPoliticoArticleUrls(x))
    return np.array(allLinks).flatten()
```

```python
def flattener(list_of_lists):    
    flattened_array = []

    for x in list_of_lists:
        for y in x:
            flattened_array.append(y)
    return flattened_array
```
