```python
def removeLessThanFourWords(text):
    regex = re.compile('[^a-zA-Z0-9 ]')
    text = re.sub(regex, '', text)
    num_words = len(text.strip().split(' '))
#     print (num_words)
#     print (text)
    return ((num_words >= 4) and ('iframe' not in text))
```

```python
# repeat the first half of the news article sentences using a function
def doubleFirstHalfArticleSentences(lst_of_sentences):
    modified = lst_of_sentences[0:math.floor(len(lst_of_sentences) / 2)] + lst_of_sentences
    return modified
```
