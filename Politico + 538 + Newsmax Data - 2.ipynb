```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyser = SentimentIntensityAnalyzer()
```

```python
def sentiment_analyzer_scores(sentence):
    score = analyser.polarity_scores(sentence)
    return score
```

```python
df['sentiment'] = df.apply(lambda row: sentiment_analyzer_scores(row['sentences']), axis=1)
```

```python
def gb_sentiment(person_flag):    # calculate daily sentiment scores for each publication for each flag
    sentiment_list = ['sentiment_compound_score', 'sentiment_pos_score', \
                      'sentiment_neu_score',  'sentiment_neg_score']
                      

    tmp_newsmax_df = df[(df[person_flag] == 1) & (df['source'] == 'newsmax')]
    tmp_newsmax_df['source'] = 'newsmax'
    
    tmp_politico_df = df[(df[person_flag] == 1) & (df['source'] == 'politico')]
    tmp_politico_df['source'] = 'politico' 
    
    merged = pd.concat([tmp_newsmax_df, tmp_politico_df])
    merged['name'] = person_flag[:-5]
    merged.reset_index(drop=True, inplace=True)
    merged_grouped = merged.groupby(['date', 'source', 'name', 'author'])[sentiment_list].mean()
    return merged_grouped
```
