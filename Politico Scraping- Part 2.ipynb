```python
def scrapePoliticoArticles(lst_of_articles):
    headers = {'User-Agent': ua.random}
    article_obj = {}
    
    for idx, link in enumerate(lst_of_articles):
        response = requests.get(link, headers)
        soup = BeautifulSoup(response.text)
        
        try:
            date_time = soup.find("time").text
        except:
            date_time = None
            
        try:
            article_title = soup.find("h1", class_=" ").text
        except:
            article_title = None
        
        try:
            author = soup.find("a", class_="url fn").text
        except:
            author = None
            
        try:
            article_site_topic = soup.find(class_="term ").text
        except:
            article_site_topic = None
            
        try:
            article_body = ''
            for tag in soup.findAll("p", class_=''):
                article_body += tag.text
        except:
            article_body = None
            
        article_obj[idx] = [date_time, article_title,author, article_site_topic, article_body]
        print(idx, "Scraped " + str(link)) 
    return article_obj
```

```python
def flattener(list_of_lists):    
    flattened_array = []

    for x in list_of_lists:
        if type(x) == str:
            flattened_array.append(x)
        else:
            for y in x:
                flattened_array.append(y)
    return flattened_array
```
